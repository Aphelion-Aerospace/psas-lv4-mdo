# Read in CFD Data

Build interpolation space in variables using [interpolate](https://docs.scipy.org/doc/scipy/reference/generated/scipy.interpolate.interp1d.html).

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
matrix = []
Dict = {}

with open('aero_database_demo.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for item in csv_reader:
        matrix.append(item)
    for k in range(len(matrix[0])):
        values = []
        for j in range(len(matrix)-1):
            value = float(matrix[j+1][k].strip())
            values.append(value)
        key = matrix[0][k].strip()
        Dict.update({key: values})
```

```python
variables = Dict.keys()
print(variables)
```

```python
zero = float(0)
for item in variables:
    Dict[item].append(zero)
    Dict[item].sort()
```

### qbar

```python
x = Dict['Mach']
xnew = np.arange(0, max(Dict['Mach']), 1e-7)
y = Dict['qbar']
f = interpolate.interp1d(x, y, kind = 'cubic')
ynew = f(xnew)   # use interpolation function returned by `interp1d`
plt.plot(x, y, 'o', xnew, ynew, '-')
plt.xlabel('Mach')
plt.ylabel('qbar')
plt.show()
```
